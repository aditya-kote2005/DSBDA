```python
import pandas as pd
```

```python
data = {
    "Name": ["A", "B", "C", "D", "E", "F"],
    "Age": [25, 30, 45, 35, 60, 28],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "Income": [30000, 40000, 55000, 48000, 70000, 42000]
}
```

```python
df = pd.DataFrame(data)
```

```python
summary = df.groupby("Gender")["Income"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)
```

```python
print(summary)
```

```python
income_mean_list = df.groupby("Gender")["Income"].mean().tolist()
print(income_mean_list)
```
