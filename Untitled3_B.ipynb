```python
import pandas as pd
```

```python
iris_data = {
    "sepal_length": [5.1, 4.9, 6.0, 5.5, 6.3, 5.8],
    "sepal_width":  [3.5, 3.0, 2.2, 2.3, 3.3, 2.7],
    "petal_length": [1.4, 1.4, 4.0, 4.5, 6.0, 5.1],
    "petal_width":  [0.2, 0.2, 1.0, 1.3, 2.5, 1.9],
    "species": [
        "Iris-setosa", "Iris-setosa",
        "Iris-versicolor", "Iris-versicolor",
        "Iris-virginica", "Iris-virginica"
    ]
}
```

```python
iris_df = pd.DataFrame(iris_data)
```

```python
species = iris_df["species"].unique()
```

```python
for sp in species:
    print("\nStatistics for:", sp)
    print(iris_df[iris_df["species"] == sp].describe())
```
